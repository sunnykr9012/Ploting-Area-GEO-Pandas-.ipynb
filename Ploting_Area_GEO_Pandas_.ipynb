```python
# Install the 'contextily' package
!pip install contextily

# Imports
import geopandas as gpd
import matplotlib.pyplot as plt
import contextily as ctx

# Load GeoDataFrames
gpd_plots = gpd.read_file("/content/plots.geojson")
gpd_BND = gpd.read_file("/content/BND_1.geojson")
gpd_canal = gpd.read_file("/content/Canal.geojson")
gpd_Roads = gpd.read_file("/content/Sherpur_Roads.geojson")
gpd_Water = gpd.read_file("/content/Sherpur_Water.geojson")

# Function to plot a multilayered map on top of a Contextily basemap.
# Accepts an arbitrary number of parameters (*args)
def makeLayeredMap(*args):
    """This function accepts an arbitrary number of geodataframes, plots them on top of a Contextily basemap.
    NOTE: Please edit the Plotting-section to specify parameters for the number of layers and the formatting of each layer.
    Output: Saved file and layered map for display."""

    # Convert the CRS for all layers to EPSG 3857 to match Contextily
    args = list(map(lambda x: x.to_crs(epsg=3857), args))

    # Create figure
    fig, ax = plt.subplots(1, figsize=(10, 10))
    ax.set_aspect('equal')

    # PLOTTING: Specify layers to plot how to format each layer (colors, transparency, etc.)
    # You can customize colors and labels for each layer here
    for i in range(len(args)):
        colors = ['blue', 'red', 'green', 'purple', 'orange']
        labels = ['Plots 1', 'BND 2', 'Water 3', 'Canal 4', 'Roads 5']
        args[i].plot(ax=ax, color=colors[i], alpha=0.5, zorder=i+1, label=labels[i])

    ctx.add_basemap(ax, source=ctx.providers.CartoDB.Positron)

    # Turn off axis
    ax.axis('off')

    # Add a legend
    ax.legend()

    # Save the map as a PDF file
    fig.savefig('layered_map.pdf', dpi=300)

    # Show the map
    plt.show()

makeLayeredMap(gpd_plots, gpd_BND, gpd_canal, gpd_Roads, gpd_Water)
```

```python
import folium
from folium import plugins
import geopandas as gpd

# Define custom style functions for each layer
def style_plots(feature):
    return {
        'fillColor': 'green',
        'color': 'black',
        'weight': 2,
    }

def style_BND(feature):
    return {
        'fillColor': 'blue',
        'color': 'black',
        'weight': 2,
    }

def style_canal(feature):
    return {
        'fillColor': 'yellow',
        'color': 'skyblue',
        'weight': 2,
    }

def style_Roads(feature):
    return {
        'color': 'brown',
        'weight': 3,
    }

def style_Water(feature):
    return {
        'fillColor': 'lightblue',
        'color': 'blue',
        'weight': 2,
    }

# Load GeoDataFrames
gpd_plots = gpd.read_file("/content/plots.geojson")
gpd_BND = gpd.read_file("/content/BND_1.geojson")
gpd_canal = gpd.read_file("/content/Canal.geojson")
gpd_Roads = gpd.read_file("/content/Sherpur_Roads.geojson")
gpd_Water = gpd.read_file("/content/Sherpur_Water.geojson")

# Create a map
m = folium.Map(location=[20.5937, 78.9629], zoom_start=5)

# Add GeoDataFrames as individual layers with custom styles
folium.GeoJson(gpd_plots, name="Plots", style_function=style_plots).add_to(m)
folium.GeoJson(gpd_BND, name="Boundaries", style_function=style_BND).add_to(m)
folium.GeoJson(gpd_canal, name="Canals", style_function=style_canal).add_to(m)
folium.GeoJson(gpd_Roads, name="Roads", style_function=style_Roads).add_to(m)
folium.GeoJson(gpd_Water, name="Water Bodies", style_function=style_Water).add_to(m)

# Add the Draw control to the map
draw = plugins.Draw(export=True)
draw.add_to(m)

# Add layer control to toggle layers on/off
folium.LayerControl().add_to(m)

# Display the map
m
```
